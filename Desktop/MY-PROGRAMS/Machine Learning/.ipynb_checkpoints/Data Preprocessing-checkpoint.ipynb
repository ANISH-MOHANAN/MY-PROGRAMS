```python
df.n_hot_rooms[(df.n_hot_rooms > 3 * uv)]=3 * uv
```

```python
df[(df.n_hot_rooms > uv)]
```

```python
np.percentile(df.rainfall,[1])[0]
```

```python
#lower value or lower variable
lv=np.percentile(df.rainfall,[1])[0]
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
clf_knn_5=KNeighborsClassifier(n_neighbors=5)
```

```python
clf_knn_5.fit(x_train_s,y_train)
```

```python
confusion_matrix(y_test,clf_knn_5.predict(x_test_s))
```

```python
accuracy_score(y_test,clf_knn_5.predict(x_test_s))
```

```python
clf_knn_7=KNeighborsClassifier(n_neighbors=7)
```

```python
clf_knn_7.fit(x_train_s,y_train)
```

```python
confusion_matrix(y_test,clf_knn_7.predict(x_test_s))
```

```python
accuracy_score(y_test,clf_knn_7.predict(x_test_s))
```

by seeing the the accuracy score of these two matrices it is clear that 7 has better accuracy
